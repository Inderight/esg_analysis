```python
num_columns=[]
for k in range(len(finan_esg.columns)):
    co_minus = [float([finan_esg.iloc[:,k][i] for i in range(1,4)][j]) >0 for j in range(3)]
    if False in co_minus :
        num_columns.append(k)
num_columns
```

```python
minus = [finan_esg.iloc[:,x].name for x in num_columns]
minus
```

```python
filter_co = finan_esg.drop(columns= minus)
filter_co.columns
```

```python
plus_co = []
for i in range(len(filter_co.columns)-1):
    if list(filter_co.columns)[i] in list(filter_co.columns)[i+1]:
        plus_co.append(list(filter_co.columns)[i+1])
plus_co
```

```python
new_plus = [plus_co[k].replace(".1","") for k in range(len(plus_co))]
new_plus
```
